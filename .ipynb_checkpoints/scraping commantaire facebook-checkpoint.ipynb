```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Initialize WebDriver with auto driver management
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()), 
    options=chrome_options
)
```

```python
url="https://www.facebook.com/MBI.NETWORK/photos/a.644456839037406/2125335377616204/"
```

```python
driver.get(url)
```

```python
data = []

a = 2
for i in range(10):
    try:
        user = driver.find_element(
            By.XPATH,
            f"/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div/div[2]/div/div/div/div[1]/div[4]/div/div[{a}]/div/div[1]/div[2]/div[1]/div[1]/div/div/span/span"
        ).text
    except Exception as e:
        user = "N/A" 
        print(f"Erreur lors de la récupération de 'user' à l'index {a}: {e}")

    try:
        commentaire = driver.find_element(
            By.XPATH,
            f"/html/body/div[1]/div/div[1]/div/div[3]/div/div/div[1]/div[1]/div/div[2]/div/div/div/div[1]/div[4]/div/div[{a}]/div/div[1]/div[2]/div[1]/div[1]/div/div/div/span/div/div"
        ).text
    except Exception as e:
        commentaire = "N/A" 
        print(f"Erreur lors de la récupération de 'commentaire' à l'index {a}: {e}")

    if user != "N/A" and commentaire != "N/A":
        data.append([user, commentaire])

    a += 1
```

```python
df = pd.DataFrame(data, columns=["User", "Commentaire"])
print(df)
```
