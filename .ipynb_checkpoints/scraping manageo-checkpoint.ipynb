```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
dt=pd.read_excel(r"entreprise2022_secteur.xlsx")
```

```python
a=dt.iloc[0:10,:]
```

```python
a
```

```python
list_url=a['url'].tolist()
```

```python
list_url
```

```python
dirgeant_list=[]
```

```python
# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Initialize WebDriver with auto driver management
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()), 
    options=chrome_options
)
```

```python
for i in list_url:
    driver.get(i)
    time.sleep(15)
    try:
        name_dirigeant = driver.find_element(By.XPATH, "/html/body/div[1]/div[3]/div/div[1]/div[2]/div/table[2]/tbody/tr[2]/td[2]")
        dirgeant_list.append(name_dirigeant.text)
        print("Dirigeant trouvé :", i)
    except NoSuchElementException:
        print("Dirigeant inexistant :", i)
        dirgeant_list.append("dirigeant_vide")
```
